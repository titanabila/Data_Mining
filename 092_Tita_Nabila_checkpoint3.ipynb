#Construct Data

###Melakukan Rekayasa Fitur

```python
def categorize_category(category):
    if 'Bar furniture' in category.lower():
        return 'Bar furnishings'
    else:
        return 'Other'

df['categoryCategory'] = df['category'].apply(categorize_category)
```

code di atas digunakan untuk membuat kolom baru dalam DataFrame yang mengkategorikan kategori berdasarkan suatu kriteria yang telah ditentukan, dalam hal ini jika kategori mengandung string 'Bar furniture', maka akan diberi label Bar furnishings, jika tidak, akan diberi label Other.
